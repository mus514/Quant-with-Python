```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
```

```python
x = np.random.normal(2.5, 1, 10000)
mu = np.linspace(2, 3, 10000)
```

```python
def ll(x, mu):
    return np.sum(norm.pdf(x = x, loc = mu))
```

```python
y = [ll(x, i) for i in mu]
```

```python
plt.scatter(mu, y)
```

```python
norm.pdf(x, 0, 1)
plt.scatter(mu, np.log(norm.pdf(x, 0, 1)))
```

```python
plt.scatter(mu, ll(mu))
```
