## Recursive Markov Chain Gambler's Ruin Game

let p probability of wining +1 and q probability of loosing -1

$$
\mu_i = P{(Ruin | X_0 = i)}
$$


$$
\mu_i = p \times \mu_{i+1} +  q \times \mu_{i-1}
$$

$$
\mu_{i+1} -\mu_i = ({\frac{q}{p}})^i \times (\mu_1 - \mu_0)
$$


$$
\mu_k = \sum_{i=0}^{k-1} ({\frac{q}{p}})^i \times (\mu_1 - \mu_0) + \mu_0
$$


$$
\mu_1 = 1 - \frac{1}{\sum_{i=0}^{k-1} ({\frac{q}{p}})^i}
$$

```python
def ruin(N, k, p):
    q = 1 - p
    if k == 0:
        return 1

    if k == N:
        return 0
    
    if k == 1:
        i = np.arange(0, N)
        x = np.sum((q / p) ** i)
        return 1 - 1 / x

    return (ruin(N, 1, p) - 1) * (q / p)**(k - 1) + ruin(N, k - 1, p)
```

```python
ruin(10, 3, 0.2)
```

## Factorial

```python
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
factorial(5) == 120
```

## Fibonacci

```python
def fib(n):
    if n < 2:
        return n
    else:
        return fib(n - 1) + fib(n - 2)
```

```python
fib(35)
```
